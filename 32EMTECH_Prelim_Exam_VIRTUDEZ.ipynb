**NUMBER 1**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

n=21

image = cv2.imread('/content/drive/MyDrive/EMTECH/circles (1).png')
cv2_imshow(image)
```

**NUMBER 2**

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_blurred = cv2.medianBlur(gray, 5, n)
```

**NUMBER 3**

```python
circles = cv2.HoughCircles(
    gray_blurred,
    cv2.HOUGH_GRADIENT,
    1,
    100,
    param1=100,
    param2=22,
    minRadius=15,
    maxRadius=300
)
```

**NUMBER 4**

```python
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        cv2.circle(image, (i[0], i[1]), i[2], (0, 0, 255), 2)
        cv2.circle(image, (i[0], i[1]), 2, (0, 0, 255), 3)
```

```python
cv2.imwrite('circles_detected.png', image)
print("Detection complete. Result saved as 'circles_detected.png'")
cv2_imshow(image)
```
